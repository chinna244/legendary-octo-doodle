```python
target = 'satisfaction'
predictors = list(df.columns) #Assigning Predictors and coverting pandas series object to list.
predictors.remove(target) #removing Target from predictors.
predictors #Checking Predictors list.
```

```python
X=df[predictors]
y=df[target]
```
